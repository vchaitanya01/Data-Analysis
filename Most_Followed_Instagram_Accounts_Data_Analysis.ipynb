# Gender Vs Followers

```python
pl = sns.countplot(x= 'Gender',data=df)
for bars in pl.containers:
    pl.bar_label(bars)
```

# Number of Followers GroupBy Gender 

```python
ins_gen = df.groupby(['Gender'], as_index=False)['Followers (Millions)'].sum().sort_values(by='Followers (Millions)', ascending=False)

ins1=sns.barplot(x = 'Gender',y= 'Followers (Millions)' ,data = ins_gen)
for bars in ins1.containers:
    ins1.bar_label(bars)
```

#  Age Group with Respect To Gender

```python
sns.countplot(data=df, x='Age_Group', hue='Gender')
```

# Followers Vs Age Group

```python
ins2 = df.groupby(['Age_Group'], as_index=False)['Followers (Millions)'].sum().sort_values(by='Followers (Millions)', ascending=False)

sns.barplot(x = 'Age_Group',y= 'Followers (Millions)' ,data = ins2)
```

Above Bars indicates that Age Group between 35-40 has more follower than other age group in Instagram

# Followers Group By Country

```python
ins3 = df.groupby(['Country'], as_index=False)['Followers (Millions)'].sum().sort_values(by='Followers (Millions)', ascending=False)

sns.barplot(x = 'Country',y= 'Followers (Millions)' ,data = ins3, width=2)
sns.set(rc={'figure.figsize':(20,3)})
```

In Above Barplot indicated United states have highest followers then other countries

# Profession Vs Followers

```python
ins5 = df.groupby(['Profession '], as_index=False)['Followers (Millions)'].sum().sort_values(by='Followers (Millions)', ascending=False)

ins1=sns.barplot(x = 'Profession ',y= 'Followers (Millions)' ,data = ins5)
```

In the above Barplot indicates that Musicians have highest Followers among all other profession
